```python
from pyarrow import parquet

tbl = parquet.read_table('../data/bger-2024-3-text.parquet')
docs = tbl['text']

extra_long_docs = []
lengths = []
for doc in docs:
    lengths.append(len(str(doc)))
    if len(str(doc)) > 100000:
        extra_long_docs.append(doc)

```

```python
import matplotlib.pyplot as plt

plt.hist([len(str(doc)) for doc in extra_long_docs], bins=30, edgecolor='black')
plt.title('Histogram of Text Lengths')
plt.xlabel('Text Length')
plt.ylabel('Frequency')
plt.show()
```
